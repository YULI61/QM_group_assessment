```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels
import statsmodels.api as sm

import numpy as np

import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('data_allclean.csv')
print(data.columns)
```

```python
data.info()
```

```python
# 

plt.figure(figsize=(15, 10))

#  `traffic_flow` 
plt.subplot(2, 3, 1)
sns.histplot(data['traffic_flow'], kde=True, bins=10,color='skyblue')
plt.title('Distribution of Traffic Flow')

#  `accident_rate` 
plt.subplot(2, 3, 2)
sns.histplot(data['accident_rate'], kde=True,bins=10, color='lightgreen')
plt.title('Distribution of Accident Rate')

#  `population_density` 
plt.subplot(2, 3, 3)
sns.histplot(data['population_density'], kde=True,bins=10, color='salmon')
plt.title('Distribution of Population Density')

#  `mean_income_of_tax_payers` 
plt.subplot(2, 3, 4)
sns.histplot(data['mean_income_of_tax_payers'], kde=True,bins=10, color='lightcoral')
plt.title('Distribution of Income of Tax Payers')

#  `Road_Network_Density` 
plt.subplot(2, 3, 5)
sns.histplot(data['Road_Network_Density'], kde=True,bins=10, color='lightblue')
plt.title('Distribution of Road Network Density')

#  `Signal_Count` 
plt.subplot(2, 3, 6)
sns.histplot(data['Signal_Count'], kde=True,bins=10, color='orange')
plt.title('Distribution of Signal Count')

# 
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

# 
plt.figure(figsize=(15, 10))

#  `traffic_flow`  Q-Q 
plt.subplot(2, 3, 1)
stats.probplot(data['traffic_flow'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Traffic Flow')

#  `accident_rate`  Q-Q 
plt.subplot(2, 3, 2)
stats.probplot(data['accident_rate'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Accident Rate')

#  `population_density`  Q-Q 
plt.subplot(2, 3, 3)
stats.probplot(data['population_density'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Population Density')

#  `mean_income_of_tax_payers`  Q-Q 
plt.subplot(2, 3, 4)
stats.probplot(data['mean_income_of_tax_payers'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Income of Tax Payers')

#  `Road_Network_Density`  Q-Q 
plt.subplot(2, 3, 5)
stats.probplot(data['Road_Network_Density'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Road Network Density')

#  `Signal_Count`  Q-Q 
plt.subplot(2, 3, 6)
stats.probplot(data['Signal_Count'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Signal Count')

# 
plt.tight_layout()
plt.show()
```
